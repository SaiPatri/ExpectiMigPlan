### 1.1 Introduction and Brief Methodology Overview

<p>
Telecommunications network provider migration is an important step and a much needed to be answered research question in both industry and academia. With the advent of new technologies and services for a host of applications, spread across multiple use-cases, user demands have increased multi-fold in a span of a few years. the onus is on network and physical infrastructure providers to upgrade their infrastructure to migrate to newer technologies. This migration is a multi-dimensional, multi-period planning problem and its solution involves forecasting, dimensioning of network infrastructure and processes, as well as evaluation of their economic costs. For optical network planners, the final technology which will provide 100 Mbps to customer is not as important as the technologies which need to be deployed and the time spent in each of these technologies, while trying to reach the final stage.
</p>
<p>
Hence, the research question we seek to answer is how can a physical infrastructure provider achieve <b>migration of all the different types of subscribers</b> (Residential, Business and ITS) to the highest data-rate, while trying to <b>maintain profitability</b>, given different <b>customer penetrations</b> and taking into account <b>customer churn</b>
</p>
<p>
To find out the best migration times and various technologies involved and trying to maximize the Net Present Value of our project in a particular scenario, we use the expectimax search approach, which is one of the earliest Artificial Intelligence decision rules algorithm and is used extensively in decision theory, game theory, statistics and deep learning. It assumes that in every move, the user (in our case, the physical infrastructure provider) will always choose to maximize their benefits and the opponent (in our case, yearly customer churn) will try to find the weighted average between the good (no churn) and bad(churn) situation, with the weights being the probability that a churn occurs.
</p>
<p>
With this expectimax search we find the expected Net Present Value in each of the four different cases for three different penetration curves (a total of 12 runs). We first choose a deployment area and calculate the number of remote nodes and lengths of ducts and fibers required for each of the 13 different technologies (varying in technology, fiber reach and datarates provided). Then we manually generate an excel file which contains the CAPEX, OPEX, Revenues and yearly cashflows (with and without customer churn), for all these technologies. This excel file is then fed as input to the expectimax search algorithm which tries to find the best possible Net Present Value, given a 10% chance of a 10% customer churn every year. The project's internal rate of return for generating the cash flows is also set to 10%. The expectimax search builds a depth first search tree and keeps adding the present values each year till it finds the maximum NPV. The corresponding path is also stored and both are displayed in the form of a graph.
</p>

### 1.4.1 Discussion of results

<p>The 4 cases remain the same, as mentioned in the previous results.</p>
<p>What is important to observe here is that the addition of business users is an advantage to the infrastructure provider since without major overhead costs, more services are added and more revenue can be generated from commercial users. However, the downside is that OPEX costs increase with the same and this is not catered to in our current OPEX model. When we add the new OPEX, these gains are bound to get smaller.
</p>
<p>
Compared to the pure residential scenario, we get better NPVs in all cases. What is also interesting to note is that most of the times, the UDWDM technology is chosen but only in the case of migrating to 100 Mbps by year 7 and having only FTTH as a provider of 100 Mbps, GPON and XGPON technology is chosen, since FTTH XGPON is cheaper to deploy than FTTH UDWDM at that particular year.
