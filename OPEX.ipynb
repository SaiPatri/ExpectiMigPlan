### 1.1 Introduction

<p>
Previously for OPEX we were using a percentage method where OPEX was calculated as 10% of Electronic Costs and 1% of non electronic costs. However, as we know OPEX are a far major factor than to be ignored, we divide the OPEX into its components and calculate the costs accordingly.
</p>
<p>
$${OPEX_b}_t = {Rent_i}_t + {Energy_i}_t + {FM_i}_t + {Marketing_i}_t + {Operations_i}_t$$
</p>
<p>
$$OPEX_i = OPEX_b*(1+\kappa)^i$$
</p>
<p></p>
<p>
where i= maxyear - currentyear, b = yearly base OPEX for each technology, t= technology, $\kappa$ = yearly inflation rate
</p>

<p>
We use the calculations for rent, energy and Failure Maintenance as given in Prof. Mas Machuca's course <i>'Techno-economic analysis for Telecommunication Networks'</i>. The calculations are done in the Excel file sent. The values for MTTR, FIT, Engineer Cost per hour are taken from the paper <i>'Cost vs. Reliability Performance Study of Fiber Access Network Architectures' by Chen et. al (link <a href="https://pdfs.semanticscholar.org/297a/704dbb0eff3f0bb1a4959d917d632fdda82c.pdf">here</a>)</i>. The energy costs and space required for each component are taken from OASE project. Finally, the rent cost per sqm is readily available on the internet from various real estate sites (approx 50€/sqm) and the cost per kWH is set to the German national average (approx 0.30€/kWH).
</p>

### 1.5 Discussion of Results

<p>We see some drastic differences in results compared to the old and the new OPEX</p>
<p>In the new OPEX, we see that the yearly costs of maintaining an FTTH network are low as compared to an FTTC or an FTTB network because we have very low energy costs and comparitively, the fault maintenance is also cheaper. This is because of the lack of DSLAM switches in the architecture which consume more energy and also fail more often. Another advantage of the FTTH implementation is that the ONTs are pushed to the customer premises, hence the PIP does not have to pay their energy costs.
</p>
<p>
Compared to the previous rather rudimentary OPEX implementation, the in-depth one leads to overall reduction in NPV costs in all the 4 different scenarios. We also see that in most the cases, the best option is to <b>DIRECTLY MIGRATE TO FTTH DWDM 100 Mbps</b> and hope to redeem the benefits over the next 20 years.
</p>


#### NOTE: Even though the year of migrations and final technologies are same in most cases, we see a variance in the NPV because of the path the migration takes i.e. the years where churn happens in each of the case is different.
